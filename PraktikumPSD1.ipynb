```python
import pandas as pd
import numpy as np
```

```python
# persiapan dataframe
n_rows = 5
n_cols = 5
cols = tuple('ABCDE')

df = pd.DataFrame(np.random.randint(1,10, size=(n_rows, n_cols)), columns=cols)
df
```

```python
df.add_suffix('field')
```

```python
#seleksi column dengan logika or
df[(df['A'] == 9)]
```

```python
#seleksi column dengan fungsi isin

df[df['A'].isin([9])]
```

```python
#memanfaatkan fungsi not dalam isin dengan operator negasi

df[~df['A'].isin([9])]
```

```python
#Konversi tipe data String ke Numerik

data = {'col 1':['1','2','3','teks'],
        'col 2':['1','2','3','4']}

df_lain = pd.DataFrame(data)
df_lain
```

```python
df_lain.dtypes # dtypes = menampilkan tipe data
```

```python
# Konversi tipe data dengan fungsi astype

df_x = df_lain.astype({'col 2': 'int'})
df_x # .astype mengubah tipe data
```

```python
df_x.dtypes
```

```python
# Konversi tipe data ke tipe data numerik

df_lain.apply(pd.to_numeric, errors='coerce')

#Coerce, kalau ada data yang tidak bisa dikonversi ke numerik, maka data tersebut akan menjadi Non
```

```python
df.index = pd.util.testing.makeDateIndex(n_rows, freq='H') #makedateIndex = data tanggal random
df = df.reset_index()

df['teks'] = list('ABCDE')

df
```

```python
# memilih kolom bertipe data spesifik
df.select_dtypes(include='number')
```
